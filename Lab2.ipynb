```python
# DO NOT MODIFY THIS CELL

from Bio import Align
from Bio.Align import substitution_matrices
from Bio.Seq import Seq
```

## Part 1: Basic Global Alignment

Use the following sequences:

- seq1 = "AAAGGTCCTAGTTGG"
- seq2 = "AAAGTCCTAGTTGG"

```python
# TODO:
# 1. Align seq1 and seq2 using global alignment and print out the alignments and their scores
# 2. Store the BEST alignment in variable: alignment

seq1 = Seq("AAAGGTCCTAGTTGG")
seq2 = Seq("AAAGTCCTAGTTGG")

aligner = Align.PairwiseAligner()
aligner.mode = 'global'

alignments = aligner.align(seq1, seq2)

for aln in alignments:
    print(f"Score: {aln.score}")
    print(aln)

alignment = alignments[0]
```

```python
### BEGIN TESTS

assert aligner is not None, "You must create an aligner"
assert alignment is not None, "You must compute an alignment"
assert hasattr(alignment, "score"), "alignment must be a Biopython alignment"

### END TESTS
print("Test passed!")
```

## Part 2: Match, Mismatch, and Gap Scoring

- match score = 2
- mismatch score = -1
- gap = -2

```python
# TODO: set match/mismatch/gap/open_gap scores, re-align seq1 and seq2, and store the score of the best alignment

aligner_mm = Align.PairwiseAligner()
aligner_mm.mode = 'global'
aligner_mm.match_score = 2
aligner_mm.mismatch_score = -1
aligner_mm.gap_score = -2

alignments_mm = aligner_mm.align(seq1, seq2)
score = alignments_mm[0].score
```

```python
### BEGIN TESTS
assert score == 26, f"Expected score 26, got {score}"
### END TESTS
print("Test passed!")
```

## Part 3: Substitution Matrix

Load the **BLOSUM62** matrix and use it in the `aligner_matrix`.

Store the best alignment in:

    matrix_alignment

```python
matrix = substitution_matrices.load("BLOSUM62")
aligner_matrix = Align.PairwiseAligner()
aligner_matrix.mode = 'global'
aligner_matrix.substitution_matrix = matrix

alignments_matrix = aligner_matrix.align(seq1, seq2)
matrix_alignment = alignments_matrix[0]
score = matrix_alignment.score
```

```python
### BEGIN TESTS

assert isinstance(matrix, substitution_matrices.Array), "Must be a Biopython substitution matrix"
assert score == 77, f"Expected score 77, got {score}"

### END TESTS
print("Test passed!")
```
